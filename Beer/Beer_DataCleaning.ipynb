```python
result = []
for f in glob.glob("Resources/*.json"):
    with open(f, "r", encoding="UTF-8") as infile:
        result.append(json.load(infile))
    with open("merged_file.json", "w", encoding="UTF-8") as outfile:
        json.dump(result, outfile)
```
